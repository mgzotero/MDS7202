```python
# TODO: cambiar prints con escritura a archivo
# TODO: graficar la V de Cramer para las variables categ√≥ricas


class Profiler:
    def __init__(self, df_in: pd.DataFrame):
        """Inicializa el profiler.

        Args:
            df_in (pd.DataFrame): DataFrame a analizar.
        """
        # Dataframe como atributo
        self.df = df_in.copy(deep=True)

        # Creaci√≥n de carpeta para guardar los resultados
        self.creation_date = pd.Timestamp.today().strftime("%d-%m-%Y")
        self.folder_name = f"EDA_{self.creation_date}"
        os.mkdir(self.folder_name)

    def summarize(self, columns: Union[str, list[str]]) -> None:
        """Genera un reporte de las columnas seleccionadas.

        Args:
            columns (Union[str, list[str]]): Columnas a analizar.
        """
        # T√≠tulo
        summary = "=== RESUMEN DE LOS DATOS ===\n"
        if isinstance(columns, str):
            columns = [columns]

        # Datatypes
        summary += "\n=== TIPOS DE VARIABLES ===\n"
        for var, dtype in self.df[columns].dtypes.astype(str).to_dict().items():
            summary += f"-- {var}: {dtype}\n"

        # Nulos
        summary += "\n=== VALORES FALTANTES ===\n"
        total_nan = self.df[columns].isna().sum().sum()
        for var, nans in self.df[columns].isna().sum().to_dict().items():
            summary += f"-- {var}: {nans} ({nans/total_nan:.2%})\n"

        # Uniques
        summary += "\n=== VALORES √öNICOS DE LAS VARIABLES CATEG√ìRICAS ===\n"
        for var, uniques in (
            self.df[columns].select_dtypes(exclude="number").nunique().to_dict().items()
        ):
            summary += f"-- {var}: {uniques} valores distintos\n"

        # Describe (num√©ricas)
        summary += "\n=== RESUMEN DE LAS VARIABLES NUM√âRICAS ===\n"
        for col in df.select_dtypes(include="number").columns:
            q1, q2, q3, q4 = df[col].quantile([0.25, 0.5, 0.75, 1])
            iqr = q3 - q1
            zero_count = df[df[col] == 0][col].count()
            neg_count = df[df[col] < 0][col].count()
            outliers_count = df[
                (df[col] < q1 - 1.5 * iqr) | (df[col] > q3 + 1.5 * iqr)
            ][col].count()
            min, max, mean = np.round((df[col].min(), df[col].max(), df[col].mean()), 2)
            summary += f"-- {col}\n"
            summary += f"Ceros: {zero_count} | Negativos: {neg_count} | Outliers: {outliers_count}\n"
            summary += f"M√≠nimo: {min} | M√°ximo: {max} | Promedio: {mean}\n"
            summary += f"Percentil 25: {q1} | Percentil 50: {q2} | Percentil 75: {q3} | Percentil 100: {q4}\n"

        # Print
        print(summary)

        # Creaci√≥n de .txt
        with open(f"{self.folder_name}/summary.txt", "w") as f:
            f.write(summary)

    def plot_vars(
        self,
        columns: Union[str, list[str]],
        n_top_cat: int = 10,
        ncols: int = 5,
        bins: int = 10,
        size: float = 0.01,
        alpha: float = 0.01,
    ) -> None:
        """ "Genera gr√°ficos de las variables del dataframe

        Args:
            columns (Union[str, list[str]]): Columnas a analizar.
        """
        # Estandarizaci√≥n del input de columnas de inter√©s
        if isinstance(columns, str):
            columns = [columns]

        # Columnas categ√≥ricas y num√©ricas
        cat_cols = self.df[columns].select_dtypes(exclude="number").columns.tolist()
        num_cols = self.df[columns].select_dtypes(include="number").columns.tolist()

        # Variables categ√≥ricas
        if len(cat_cols) > 0:
            print("=== VARIABLES CATEG√ìRICAS ===")
            for col in cat_cols:
                fig, ax = plt.subplots()
                series = self.df.select_dtypes(exclude="number")[col].value_counts()
                series = series[:n_top_cat]
                ax.bar(series.index, series.values)
                ax.set_title(col)
                ax.set_xticks(
                    ax.get_xticks(), ax.get_xticklabels(), rotation=45, ha="right"
                )
                fig.tight_layout()
                fig.savefig(f"{self.folder_name}/{col}.pdf")
            # TODO: V de Cr√°mer
        else:
            print("No hay variables categ√≥ricas")
        
        # Variables num√©ricas
        if len(num_cols) > 0:
            ...
        else:
            print("No hay variables num√©ricas")
        
        def clean_data(self):
            ...
        
        def scale(self):
            ...

        def make_clusters(self):
            ...
        
        def detect_anomalies(self):
            ...
        
        def profile(self):
            

        
```
