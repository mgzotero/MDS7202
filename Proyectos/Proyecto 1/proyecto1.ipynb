```python
class Profiler:
    def __init__(self, df_in: pd.DataFrame):
        """Inicializa el profiler.

        Args:
            df_in (pd.DataFrame): DataFrame a analizar.
        """
        # Dataframes como atributos
        self.df = df_in.copy(deep=True)  # original
        self.clean_df = None  # clean
        self.scaled_df = None  # scaled
        self.cluster_df = None  # clusters
        self.anomaly_df = None  # anomal√≠as

        # Creaci√≥n de carpeta para guardar los resultados
        self.creation_date = pd.Timestamp.today().strftime("%d-%m-%Y")
        self.folder_name = f"EDA_{self.creation_date}"
        try:
            shutil.rmtree(self.folder_name)
        except:
            pass
        os.mkdir(self.folder_name)

    def summarize(
        self,
        columns: Union[str, list[str]],
        df: str,
        nan_cutoff: float = 0.5,
        show: bool = True,
    ) -> None:
        """Genera un reporte de las columnas seleccionadas.

        Args:
            columns (Union[str, list[str]]): Columnas a analizar.
            df (str): DataFrame a utilizar (original, clean, scaled, clusters, anomalies).
            nan_cutoff (float, opcional): Porcentaje de nulos a partir del cual se genera una alerta.
            show (bool, opcional): Mostrar el reporte.
        """
        # Estandarizaci√≥n del input de columnas de inter√©s
        if isinstance(columns, str):
            columns = [columns]

        # Elegimos el dataframe a utilizar
        df_dict = {
            "original": self.df,
            "clean": self.clean_df,
            "scaled": self.scaled_df,
            "clusters": self.cluster_df,
            "anomalies": self.anomaly_df,
        }
        try:
            df = df_dict[df][columns].copy(deep=True)
        except:
            raise ValueError("El dataframe especificado no existe")
        del df_dict

        # T√≠tulo
        summary = "=== RESUMEN DE LOS DATOS ===\n\n\n"

        # Datatypes
        summary += "\n=== TIPOS DE VARIABLES ===\n"
        for var, dtype in df[columns].dtypes.astype(str).to_dict().items():
            summary += f"-- {var}: {dtype}\n"

        # Nulos
        summary += "\n=== VALORES FALTANTES ===\n"
        for var, nans in df[columns].isna().sum().to_dict().items():
            summary += f"-- {var}: {nans} ({nans/df.shape[0]:.2%})\n"

        # Uniques
        summary += "\n=== VALORES √öNICOS DE LAS VARIABLES CATEG√ìRICAS ===\n"
        for var, uniques in (
            df[columns].select_dtypes(exclude="number").nunique().to_dict().items()
        ):
            summary += f"-- {var}: {uniques} valores distintos\n"

        # Describe (num√©ricas)
        summary += "\n=== RESUMEN DE LAS VARIABLES NUM√âRICAS ===\n"
        for col in df.select_dtypes(include="number").columns:
            q1, q2, q3, q4 = df[col].quantile([0.25, 0.5, 0.75, 1])
            iqr = q3 - q1  # m√©todo intercuartil
            zero_count = df[df[col] == 0][col].count()
            neg_count = df[df[col] < 0][col].count()
            outliers_count = df[
                (df[col] < q1 - 1.5 * iqr) | (df[col] > q3 + 1.5 * iqr)
            ][col].count()
            min, max, mean = np.round((df[col].min(), df[col].max(), df[col].mean()), 2)
            summary += f"-- {col}\n"
            summary += f"Ceros: {zero_count} | Negativos: {neg_count} | Outliers (IQR): {outliers_count}\n"
            summary += f"M√≠nimo: {min} | M√°ximo: {max} | Promedio: {mean}\n"
            summary += f"Percentil 25: {q1} | Percentil 50: {q2} | Percentil 75: {q3} | Percentil 100: {q4}\n"

        # Alerta de nulos
        nans = df.isna().sum() / df.shape[0] * 100
        cols_above_cutoff = nans[nans > nan_cutoff].index.tolist()
        if len(cols_above_cutoff) > 0:
            summary += "\n\n\n=== ALERTA ===\n"
            summary += f"Las siguientes columnas sobrepasan el l√≠mite de porcentaje de nulos ({nan_cutoff*100:.2f}%):\n"
            for col in cols_above_cutoff:
                summary += f"-- {col}\n"

        # Display
        print(summary) if show else None

        # Creaci√≥n de .txt
        try:
            shutil.rmtree(f"{self.folder_name}/summary.txt")
        except:
            pass
        with open(f"{self.folder_name}/summary.txt", "w") as f:
            f.write(summary)

    def plot_vars(
        self,
        columns: Union[str, list[str]],
        df: str,
        show: bool = True,
        n_top_cat: int = 10,
    ) -> None:
        """ "Genera gr√°ficos de las variables del dataframe

        Args:
            columns (Union[str, list[str]]): Columnas a analizar.
            show (bool, opcional): Mostrar los gr√°ficos.
            n_top_cat (int, opcional): N√∫mero de categor√≠as a mostrar en los gr√°ficos de barras.
        """
        # Estandarizaci√≥n del input de columnas de inter√©s
        if isinstance(columns, str):
            columns = [columns]

        # Elegimos el dataframe a utilizar
        df_dict = {
            "original": self.df,
            "clean": self.clean_df,
            "scaled": self.scaled_df,
            "clusters": self.cluster_df,
            "anomalies": self.anomaly_df,
        }
        try:
            df = df_dict[df][columns].copy(deep=True)
        except:
            raise ValueError("El dataframe especificado no existe")
        del df_dict

        # Columnas categ√≥ricas y num√©ricas
        cat_cols = df[columns].select_dtypes(exclude="number").columns.tolist()
        num_cols = df[columns].select_dtypes(include="number").columns.tolist()

        # Creaci√≥n de carpeta para guardar los plots
        try:
            shutil.rmtree(f"{self.folder_name}/plots")
        except:
            pass
        os.mkdir(f"{self.folder_name}/plots")

        ## Variables categ√≥ricas (univariado)
        if len(cat_cols) > 0:
            print("=== VARIABLES CATEG√ìRICAS (UNIVARIADO) ===") if show else None
            for col in cat_cols:
                fig, ax = plt.subplots()
                series = df.select_dtypes(exclude="number")[col].value_counts()
                n_top_cat_min = np.min(
                    [n_top_cat, len(series)]
                )  # m√≠nimo entre el n√∫mero de categor√≠as y el l√≠mite
                series = series[
                    :n_top_cat_min
                ]  # seleccionamos m√°ximo n_top_cat categor√≠as
                ax.bar(series.index, series.values)
                title = (
                    f"Categor√≠as de {col}"
                    if n_top_cat_min != n_top_cat
                    else f"Top {n_top_cat} categor√≠as de {col}"
                )
                ax.set_title(title)
                ax.set_xticks(
                    ax.get_xticks(), ax.get_xticklabels(), rotation=45, ha="right"
                )
                ax.set_ylabel("Frecuencia")
                fig.tight_layout()
                plt.show() if show else plt.close()
                fig.savefig(f"{self.folder_name}/plots/{col}.pdf")  # guardamos el plot

            ## Variables categ√≥ricas (multivariado)
            if len(cat_cols) > 1:
                print("=== VARIABLES CATEG√ìRICAS (MULTIVARIADO) ===") if show else None
                cramers_matrix = np.zeros((len(cat_cols), len(cat_cols)))
                for i in range(len(cat_cols) - 1):
                    for j in range(i + 1, len(cat_cols)):
                        try:
                            contingency_table = pd.crosstab(
                                df[cat_cols[i]], df[cat_cols[j]]
                            )
                            cramers_v = association(contingency_table, method="cramer")
                            cramers_matrix[i, j] = cramers_v
                            cramers_matrix[j, i] = cramers_v
                        except:
                            # Si la tabla de contingencia es demasiado grande, se obtiene un error de memoria
                            cramers_matrix[
                                i, j
                            ] = -1  # se asigna -1 para identificar que hubo un error
                            cramers_matrix[j, i] = -1
                fig, ax = plt.subplots(figsize=(10, 8))
                sns.heatmap(
                    cramers_matrix,
                    annot=True,
                    xticklabels=cat_cols,
                    yticklabels=cat_cols,
                    ax=ax,
                )
                ax.set_title("Correlaci√≥n entre variables categ√≥ricas (V de Cr√°mer)")
                fig.tight_layout()
                plt.show() if show else plt.close()
                fig.savefig(f"{self.folder_name}/plots/v-de-cramer.pdf")
        else:
            print("No hay variables categ√≥ricas")

        ## Variables num√©ricas (univariado)
        if len(num_cols) > 0:
            print("=== VARIABLES NUM√âRICAS (UNIVARIADO) ===") if show else None
            for col in num_cols:
                fig, ax = plt.subplots()
                sns.kdeplot(df[col], ax=ax)  # densidad
                ax.set_title(f"Densidad de {col}")
                ax.set_ylabel("Densidad")
                fig.tight_layout()
                plt.show() if show else plt.close()
                fig.savefig(f"{self.folder_name}/plots/{col}.pdf")

            ## Variables num√©ricas (multivariado)
            if len(num_cols) > 1:
                print("=== VARIABLES NUM√âRICAS (MULTIVARIADO) ===") if show else None
                fig, ax = plt.subplots()
                sns.heatmap(df[num_cols].corr(), annot=True, ax=ax)
                ax.set_title("Correlaci√≥n entre variables num√©ricas")
                fig.tight_layout()
                plt.show() if show else plt.close()
                fig.savefig(f"{self.folder_name}/plots/correlation-matrix.pdf")
        else:
            print("No hay variables num√©ricas")

    def clean_data(
        self,
        columns: Union[str, list[str]],
        nan_technique: str,
        split_non_atomic: dict[str, list[str]],
        custom_nan_replacement: dict[str, str],
        nan_str: str,
    ) -> None:
        """Limpia los datos de acuerdo a los criterios especificados.

        Args:
            columns (Union[str, list[str]]): Columnas a analizar.
            nan_technique (str, opcional): T√©cnica de tratamiento de nulos (drop, imputation).
            split_non_atomic (dict[str, list[str]]): Columna(s) no at√≥mica(s) a separar.
            custom_nan_replacement (dict[str, str]): Reemplazo de nulos espec√≠fico.
            nan_str (str): String que representa un nulo.
        """
        # Creaci√≥n de carpeta para guardar los datos limpios
        try:
            shutil.rmtree(f"{self.folder_name}/clean_data")
        except:
            pass
        os.mkdir(f"{self.folder_name}/clean_data")

        # Estandarizaci√≥n del input de columnas de inter√©s
        if isinstance(columns, str):
            columns = [columns]

        # Copia del dataframe
        clean_df = self.df[columns].copy(deep=True)

        ## Columnas categ√≥ricas y num√©ricas
        # cat_cols = clean_df[columns].select_dtypes(exclude="number").columns.tolist()
        # num_cols = clean_df[columns].select_dtypes(include="number").columns.tolist()

        # Funci√≥n de transformaci√≥n
        def transformer(df: pd.DataFrame) -> pd.DataFrame:
            """Transforma el dataframe de acuerdo a los criterios especificados.

            Args:
                df (pd.DataFrame): DataFrame a transformar.

            Returns:
                pd.DataFrame: DataFrame transformado.
            """
            # Columna(s) no at√≥mica(s)
            if len(split_non_atomic) > 0:
                delimiters = [punc for punc in string.punctuation if punc != "."]
                for (
                    col,
                    subcols,
                ) in (
                    split_non_atomic.items()
                ):  # se recorren las distintas columnas no at√≥micas
                    splitcol = df[col].copy(deep=True)  # serie temporal
                    for (
                        delim
                    ) in delimiters:  # se reemplazan los delimitadores por espacios
                        splitcol = splitcol.str.replace(delim, " ")
                    try:
                        df[subcols] = (
                            splitcol.str.split(" ", expand=True)
                            .replace(nan_str, np.nan)
                            .astype(float)
                        )
                    except:
                        df[subcols] = splitcol.str.split(" ", expand=True).replace(
                            nan_str, np.nan
                        )
                    df.drop(columns=col, inplace=True)

            # Columnas categ√≥ricas y num√©ricas despu√©s de la separaci√≥n
            cat_cols = df.select_dtypes(exclude="number").columns.tolist()
            num_cols = df.select_dtypes(include="number").columns.tolist()

            # Tratamiento de nulos (espec√≠fico)
            if len(custom_nan_replacement) > 0:
                for col, nan_replacement in custom_nan_replacement.items():
                    df[col] = df[col].fillna(nan_replacement)

            # Tratamiento de nulos (general)
            if nan_technique == "drop":
                df = df.dropna()
            elif nan_technique == "imputation":
                df[num_cols] = (
                    df[num_cols].fillna(df[num_cols].mean().iloc[0]).copy(deep=True)
                )  # columnas num√©ricas
                df[cat_cols] = (
                    df[cat_cols].fillna(df[cat_cols].mode().iloc[0]).copy(deep=True)
                )  # columnas categ√≥ricas

            # Dropeamos duplicados
            df = df.drop_duplicates()

            # Eliminamos outliers
            for col in num_cols:
                q1, q3 = df[col].quantile([0.25, 0.75])
                iqr = q3 - q1
                df = df[(df[col] >= q1 - 1.5 * iqr) & (df[col] <= q3 + 1.5 * iqr)].copy(
                    deep=True
                )

            return df

        # Instanciamos el transformer
        ft = FunctionTransformer(transformer).set_output(transform="pandas")

        # Transformamos el dataframe y lo guardamos
        clean_df = ft.transform(clean_df)
        clean_df.to_csv(f"{self.folder_name}/clean_data/clean_data.csv", index=False)
        self.clean_df = clean_df.copy(deep=True)

        # Mensaje de confirmaci√≥n
        print(
            "Dataframe limpio guardado en la carpeta clean_data y como nuevo atributo '.clean_df'",
            "(para acceder al dataframe original, usar el atributo '.original_df')",
            sep="\n",
            end="\n",
        )

    def scale(
        self,
        columns: Union[str, list[str]],
        num_func_transformer: Callable,
        num_scaler: Any,
        cat_encoder: Any,
    ) -> None:
        """Escala las variables num√©ricas y transforma las variables categ√≥ricas.

        Args:
            num_func_transformer (function): Funci√≥n de transformaci√≥n de las variables num√©ricas.
            num_scaler (Any): Scaler de las variables num√©ricas.
            cat_encoder (Any, opcional): Encoder de las variables categ√≥ricas.
        """
        # Estandarizaci√≥n del input de columnas de inter√©s
        if isinstance(columns, str):
            columns = [columns]

        # Copia del dataframe
        df = self.clean_df[columns].copy(deep=True)

        # Creaci√≥n de carpeta para guardar los datos escalados
        try:
            shutil.rmtree(f"{self.folder_name}/scale")
        except:
            pass
        os.mkdir(f"{self.folder_name}/scale")

        # Columnas categ√≥ricas y num√©ricas
        cat_cols = df.select_dtypes(exclude="number").columns.tolist()
        num_cols = df.select_dtypes(include="number").columns.tolist()

        # Transformaci√≥n de variables categ√≥ricas
        cat_transformer = Pipeline(steps=[("encoder", cat_encoder)])

        # Transformaci√≥n de variables num√©ricas
        num_steps = list()
        if num_func_transformer is not None:
            num_steps.append(
                ("func_transformer", FunctionTransformer(num_func_transformer))
            )
        if num_scaler is not None:
            num_steps.append(("scaler", num_scaler))
        num_transformer = Pipeline(steps=num_steps)

        # ColumnTransformer
        ct = ColumnTransformer(
            transformers=[
                ("cat", cat_transformer, cat_cols),
                ("num", num_transformer, num_cols),
            ]
        )

        # Transformamos el dataframe y lo guardamos
        scaled_df = ct.fit_transform(df)

        # Recuperamos los nombres de las columnas
        cat_cols = ct.named_transformers_["cat"].get_feature_names_out().tolist()
        cols = cat_cols + num_cols

        # Creamos el dataframe
        scaled_df = pd.DataFrame(scaled_df, columns=cols)
        scaled_df.to_csv(f"{self.folder_name}/scale/scaled_features.csv", index=False)

        # Guardamos los datos escalados como atributo
        self.scaled_df = scaled_df.copy(deep=True)

        # Mensaje de confirmaci√≥n
        print(
            "Dataframe escalado guardado en la carpeta scale y como nuevo atributo '.scaled_df'",
            sep="\n",
            end="\n",
        )

    def make_clusters(
        self,
        columns: Union[str, list[str]],
        clustering_model: Any,
        max_clusters: int = 10,
        df: str = "scaled",
        viz_frac: float = 0.1,
        dim_reduc_model: Any = TSNE(n_components=2, random_state=42),
        show: bool = True,
        random_state: int = 42,
        nan_technique: str = "imputation",
        split_non_atomic: dict[str, list[str]] = {
            "age-height-weight": ["Age", "Height", "Weight"]
        },
        custom_nan_replacement: dict[str, str] = {"Medal": "No_Medal"},
        nan_str: str = "nan",
        num_func_transformer: Callable = np.log,
        num_scaler: Any = MinMaxScaler(),
        cat_encoder: Any = OneHotEncoder(handle_unknown="ignore"),
    ) -> None:
        """Genera clusters de los datos escalados.

        Args:
            columns (Union[str, list[str]]): Columnas a analizar.
            clustering_model (Any): Modelo de clustering.
            max_clusters (int, opcional): N√∫mero m√°ximo de clusters a generar.
            df (str, opcional): DataFrame a utilizar (original, clean, scaled).
            viz_frac (float, opcional): Fracci√≥n de datos a samplear para visualizaci√≥n.
            dim_reduc_model (Any, opcional): Modelo de reducci√≥n de dimensionalidad.
            show (bool, opcional): Mostrar los gr√°ficos.
            random_state (int, opcional): Semilla para reproducibilidad.
            nan_technique (str, opcional): T√©cnica de tratamiento de nulos (drop, imputation).
            split_non_atomic (dict[str, list[str]], opcional): Columna(s) no at√≥mica(s) a separar.
            custom_nan_replacement (dict[str, str], opcional): Reemplazo de nulos espec√≠fico.
            nan_str (str, opcional): String que representa un nulo.
            num_func_transformer (Callable, opcional): Funci√≥n de transformaci√≥n de las variables num√©ricas.
            num_scaler (Any, opcional): Scaler de las variables num√©ricas.
            cat_encoder (Any, opcional): Encoder de las variables categ√≥ricas.
        """
        # Estandarizaci√≥n del input de columnas de inter√©s
        if isinstance(columns, str):
            columns = [columns]

        # Creaci√≥n de carpeta para guardar los clusters
        try:
            shutil.rmtree(f"{self.folder_name}/clusters")
        except:
            pass
        os.mkdir(f"{self.folder_name}/clusters")

        # Elegimos el dataframe a utilizar
        if df != "preprocess":
            df_dict = {
                "original": self.df,
                "clean": self.clean_df,
                "scaled": self.scaled_df,
                "clusters": self.cluster_df,
                "anomalies": self.anomaly_df,
            }
            try:
                df = df_dict[df][columns].copy(deep=True)
            except:
                raise ValueError("El dataframe especificado no existe")
            del df_dict
        else:
            self.clean_data(
                columns,
                nan_technique,
                split_non_atomic,
                custom_nan_replacement,
                nan_str,
            )
            self.scale(
                columns,
                num_func_transformer,
                num_scaler,
                cat_encoder,
            )
            df = self.scaled_df.copy(deep=True)

        # M√©todo del codo
        if "n_clusters" in clustering_model.get_params().keys():
            wcss = list()  # within cluster sum of squares
            for k in range(1, max_clusters + 1):
                clustering_model.set_params(n_clusters=k)
                clustering_model.fit(df)
                wcss.append(clustering_model.inertia_)

            # Gr√°fico del m√©todo del codo
            plt.plot(list(range(1, len(wcss) + 1)), wcss)
            plt.xticks(list(range(1, len(wcss) + 1)))
            plt.title("M√©todo del codo")
            plt.savefig(f"{self.folder_name}/clusters/metodo-del-codo.pdf")
            plt.show()

            # Elegimos el n√∫mero √≥ptimo de clusters
            k_opt = int(
                input("Dado el gr√°fico, ¬øcu√°l es el n√∫mero √≥ptimo de clusters?")
            )
            print(
                f"Modelo de clustering elegido: {type(clustering_model).__name__} (n_clusters = {k_opt})"
            )
        else:
            print("Modelo de clustering no admite par√°metro 'n_clusters'")

        # Entrenamos el modelo
        if "n_clusters" in clustering_model.get_params().keys():
            clustering_model.set_params(n_clusters=k_opt)
        clustering_model.fit(df)

        # Predecimos los clusters
        clust_pred = clustering_model.predict(df)
        df["cluster"] = clust_pred

        # Guardamos los clusters
        df.to_csv(f"{self.folder_name}/clusters/clusters.csv", index=False)
        self.df_clusters = df.copy(deep=True)

        # Sampleo para visualizaci√≥n
        # if df.shape[0] > 1000:
        #     idx = df.sample(frac=viz_frac, random_state=random_state).index
        #     df = df.loc[idx, :]

        # Reducci√≥n de dimensionalidad
        df_dimred = pd.DataFrame(
            dim_reduc_model.fit_transform(df.drop(columns="cluster")),
            columns=["x1", "x2"],
        )
        df_dimred["cluster"] = df["cluster"].copy(deep=True)

        # Visualizaci√≥n
        if show:
            fig, ax = plt.subplots()
            sns.scatterplot(data=df_dimred, x="x1", y="x2", hue="cluster", ax=ax)
            ax.set_title("Visualizaci√≥n de los clusters")
            fig.tight_layout()
            plt.savefig(f"{self.folder_name}/clusters/plot-clusters.pdf")
            plt.show()

    def detect_anomalies(
        self,
        columns: Union[str, list[str]],
        anomaly_detector: Any,
        contamination: float = 0.05,
        df: str = "scaled",
        random_state: int = 42,
        viz_frac: float = 0.1,
        show: bool = True,
        dim_reduc_model: Any = TSNE(n_components=2, random_state=42),
        nan_technique: str = "imputation",
        split_non_atomic: dict[str, list[str]] = {
            "age-height-weight": ["Age", "Height", "Weight"]
        },
        custom_nan_replacement: dict[str, str] = {"Medal": "No_Medal"},
        nan_str: str = "nan",
        num_func_transformer: Callable = np.log,
        num_scaler: Any = MinMaxScaler(),
        cat_encoder: Any = OneHotEncoder(handle_unknown="ignore"),
    ) -> None:
        """Detecta anomal√≠as en los datos.

        Args:
            columns (Union[str, list[str]]): Columnas a analizar.
            anomaly_detector (Any): Detector de anomal√≠as.
            contamination (float, opcional): Porcentaje de anomal√≠as esperadas.
            df (str, opcional): DataFrame a utilizar (original, clean, scaled).
            random_state (int, opcional): Semilla para reproducibilidad.
            viz_frac (float, opcional): Fracci√≥n de datos a samplear para visualizaci√≥n.
            dim_reduc_model (Any, opcional): Modelo de reducci√≥n de dimensionalidad.
            nan_technique (str, opcional): T√©cnica de tratamiento de nulos (drop, imputation).
            split_non_atomic (dict[str, list[str]], opcional): Columna(s) no at√≥mica(s) a separar.
            custom_nan_replacement (dict[str, str], opcional): Reemplazo de nulos espec√≠fico.
            nan_str (str, opcional): String que representa un nulo.
            num_func_transformer (Callable, opcional): Funci√≥n de transformaci√≥n de las variables num√©ricas.
            num_scaler (Any, opcional): Scaler de las variables num√©ricas.
            cat_encoder (Any, opcional): Encoder de las variables categ√≥ricas.
        """
        # Estandarizaci√≥n del input de columnas de inter√©s
        if isinstance(columns, str):
            columns = [columns]

        # Creaci√≥n de carpeta para guardar los resultados
        try:
            shutil.rmtree(f"{self.folder_name}/anomalies")
        except:
            pass
        os.mkdir(f"{self.folder_name}/anomalies")

        # Elegimos el dataframe a utilizar
        if df != "preprocess":
            df_dict = {
                "original": self.df,
                "clean": self.clean_df,
                "scaled": self.scaled_df,
                "clusters": self.cluster_df,
                "anomalies": self.anomaly_df,
            }
            try:
                df = df_dict[df][columns].copy(deep=True)
            except:
                raise ValueError("El dataframe especificado no existe")
            del df_dict
        else:
            self.clean_data(
                columns,
                nan_technique,
                split_non_atomic,
                custom_nan_replacement,
                nan_str,
            )
            self.scale(
                columns,
                num_func_transformer,
                num_scaler,
                cat_encoder,
            )
            df = self.scaled_df.copy(deep=True)

        # Entrenamos el modelo
        try:
            anomaly_detector.set_params(contamination=contamination)
        except:
            pass
        anomaly_detector.fit(df)

        # Predecimos las anomal√≠as
        anomaly_pred = anomaly_detector.predict(df)
        df["anomaly"] = anomaly_pred
        df["anomaly"] = df["anomaly"].map(lambda x: 1 if x == -1 else 0)

        # Guardamos los resultados
        df.to_csv(f"{self.folder_name}/anomalies/anomalies.csv", index=False)
        self.df_anomalies = df.copy(deep=True)

        # Sampleo para visualizaci√≥n
        # if df.shape[0] > 1000:
        #     idx = df.sample(frac=viz_frac, random_state=random_state).index
        #     df = df.loc[idx, :]

        # Reducci√≥n de dimensionalidad
        df_dimred = pd.DataFrame(
            dim_reduc_model.fit_transform(df.drop(columns="anomaly")),
            columns=["x1", "x2"],
        )
        df_dimred["anomaly"] = df["anomaly"].copy(deep=True)

        # Visualizaci√≥n
        if show:
            fig, ax = plt.subplots()
            sns.scatterplot(data=df_dimred, x="x1", y="x2", hue="anomaly", ax=ax)
            ax.set_title("Visualizaci√≥n de las anomal√≠as")
            fig.tight_layout()
            plt.savefig(f"{self.folder_name}/anomalies/plot-anomalies.pdf")
            plt.show()

    def profile(
        self,
        columns: Union[str, list[str]],
        scale_cols: Union[str, list[str]] = None,
        df_summarize: str = "clean",
        df_plot: str = "clean",
        df_clusters: str = "scaled",
        df_anomalies: str = "scaled",
        nan_cutoff: float = 0.5,
        n_top_cat: int = 10,
        show: bool = True,
        nan_technique: str = "imputation",
        split_non_atomic: dict[str, list[str]] = {
            "age-height-weight": ["Age", "Height", "Weight"]
        },
        custom_nan_replacement: dict[str, str] = {"Medal": "No_Medal"},
        nan_str: str = "nan",
        num_func_transformer: Callable = np.log,
        num_scaler: Any = MinMaxScaler(),
        cat_encoder: Any = OneHotEncoder(handle_unknown="ignore"),
        clustering_model: Any = KMeans(),
        max_clusters: int = 10,
        viz_frac: float = 0.1,
        random_state: int = 42,
        anomaly_detector: Any = IsolationForest(),
        contamination: float = 0.05,
    ) -> None:
        """Ejecuta todos los m√©todos del profiler.

        Args:
            columns (Union[str, list[str]]): Columnas a analizar.
            scale_cols (Union[str, list[str]], opcional): Columnas a escalar.
            nan_cutoff (float, opcional): Porcentaje de nulos a partir del cual se genera una alerta.
            n_top_cat (int, opcional): N√∫mero de categor√≠as a mostrar en los gr√°ficos de barras.
            show (bool, opcional): Mostrar los gr√°ficos.
            nan_technique (str, opcional): T√©cnica de tratamiento de nulos (drop, imputation).
            split_non_atomic (dict[str, list[str]], opcional): Columna(s) no at√≥mica(s) a separar.
            custom_nan_replacement (dict[str, str], opcional): Reemplazo de nulos espec√≠fico.
            nan_str (str, opcional): String que representa un nulo.
            num_func_transformer (Callable, opcional): Funci√≥n de transformaci√≥n de las variables num√©ricas.
            num_scaler (Any, opcional): Scaler de las variables num√©ricas.
            cat_encoder (Any, opcional): Encoder de las variables categ√≥ricas.
            clustering_model (Any, opcional): Modelo de clustering.
            max_clusters (int, opcional): N√∫mero m√°ximo de clusters a generar.
            viz_frac (float, opcional): Fracci√≥n de datos a samplear para visualizaci√≥n.
            random_state (int, opcional): Semilla para reproducibilidad.
            anomaly_detector (Any, opcional): Detector de anomal√≠as.
            contamination (float, opcional): Porcentaje de anomal√≠as esperadas.
        """
        # Estandarizaci√≥n del input de columnas de inter√©s
        if isinstance(columns, str):
            columns = [columns]

        # Limpieza de datos
        self.clean_data(
            columns,
            nan_technique,
            split_non_atomic,
            custom_nan_replacement,
            nan_str,
        )

        # Resumen
        self.summarize(columns, df_summarize, nan_cutoff, show)

        # Gr√°ficos
        self.plot_vars(columns, df_plot, show, n_top_cat)

        # Columnas para escalar
        if scale_cols is None:
            scale_cols = columns
        columns = scale_cols

        # Escalado
        self.scale(
            columns,
            num_func_transformer=num_func_transformer,
            num_scaler=num_scaler,
            cat_encoder=cat_encoder,
        )

        # Columnas nuevas
        columns = columns + cat_encoder.get_feature_names_out().tolist()

        # Clusters
        self.make_clusters(
            columns,
            clustering_model=clustering_model,
            max_clusters=max_clusters,
            df=df_clusters,
            viz_frac=viz_frac,
            random_state=random_state,
        )

        # Anomal√≠as
        self.detect_anomalies(
            columns,
            anomaly_detector=anomaly_detector,
            contamination=contamination,
            df=df_anomalies,
            random_state=random_state,
            viz_frac=viz_frac,
        )

    def clearGarbage(self) -> None:
        """Elimina la carpeta creada para guardar los resultados."""
        shutil.rmtree(self.folder_name)
```

A continuaci√≥n se proceder√° a graficar la distribuci√≥n e interacciones de las variables del dataset. Esto se realizar√° utilizando el m√©todo `plot_vars` de la clase `Profiler`.

```python
pf.plot_vars(
    df="original",  # original, clean, scaled, clusters, anomalies
    columns=cols,  # columnas a analizar
    n_top_cat=10,  # n√∫mero de categor√≠as a mostrar en los gr√°ficos de barras
    show=True,  # mostrar los gr√°ficos
 )
```

En los gr√°ficos presentados es posible observar la manifestaci√≥n de los problemas mencionados, principalmente aquellos derivados de la columna no at√≥mica, pues la matriz de V de Cr√°mer se ve obligada a imponer un `-1` en el cruce de `age-height-weight` con la columna `Name`, pues surge un error de falta de memoria al ejecutar el `crosstab` (matriz de contingencia). Otro aspecto a considerar, es que al ser una columna no at√≥mica que encapsula variables num√©ricas, se pierde informaci√≥n dentro de la matriz de correlaci√≥n. En cuanto a la matriz de V de Cr√°mer en s√≠, es posible observar que existen bastantes variables categ√≥ricas que presentan una correlaci√≥n significativa con las dem√°s variables categ√≥ricas (e.g. City-Event-Sport-Season).

Para enfrentar los problemas mencionados anteriormente, se proceder√° a limpiar los datos utilizando el m√©todo `clean_data` de la clase `Profiler`. Este m√©todo se encarga de limpiar los datos para que luego puedan ser procesados. Para esto, se crear√° una carpeta `EDA_fecha/clean_data` y se implementar√° una funcionalidad para filtrar y aplicar este m√©todo a una o m√°s variables de inter√©s. En este caso, se aplicar√° a todas las variables del dataset.

```python
pf.clean_data(
    columns=cols,  # columnas a analizar
    split_non_atomic={"age-height-weight": ["Age", "Height", "Weight"]},  # separamos la columna "age-height-weight"
    custom_nan_replacement={"Medal": "No_Medal"},  # imputamos con "No_Medal"
    nan_str="nan",  # especificamos que los nulos est√°n representados por "nan" en la columna no at√≥mica
    nan_technique="imputation",  # imputamos los nulos
)
```

```python
pf.clean_df.head()
```

Tras haber limpiado los datos, se proceder√° a resuimir y graficar nuevamente la distribuci√≥n e interacciones de las variables del dataset. Esto se realizar√° utilizando los m√©todos `summarize` y `plot_vars` de la clase `Profiler`.

```python
cols = pf.clean_df.columns.tolist()  # columnas a analizar
pf.summarize(
    df="clean",  # original, clean, scaled, clusters, anomalies
    columns=cols,  # lista de columnas
    nan_cutoff=0.30,  # porcentaje de nulos a partir del cual se genera una alerta
    show=True,  # mostrar el reporte
)
```

```python
pf.plot_vars(
    df="clean",  # original, clean, scaled, clusters, anomalies
    columns=cols,  # columnas a analizar
    n_top_cat=10,  # n√∫mero de categor√≠as a mostrar en los gr√°ficos de barras
    show=True,  # mostrar los gr√°ficos
)
```

Es posible notar que al separar la columna no at√≥mica en sus respectivas columnas, se logra una mejor visualizaci√≥n de la matriz de correlaci√≥n, pues se logra observar correlaciones interesantes entre la edad, el peso y la altura (v√©ase la correlaci√≥n entre "weight" y "height"). Adem√°s es posible observar que las variables "age", "height" y "weight" adquieren distribuciones que se asemejan a una normal.

Adem√°s, se logra observar que la columna `Medal` sufr√≠a de una gran cantidad de valores nulos debido a que no todos los participantes obtienen medallas en los juegos ol√≠mpicos. Esto se puede observar en el gr√°fico de barras de la columna `Medal`, donde se puede apreciar que la mayor√≠a de los participantes no obtienen medallas.

A continuaci√≥n escalaremos los datos utilizando un subconjunto de las features provistas, esto debido a que existen variables categ√≥ricas con muchos valores disintos, lo que podr√≠a generar problemas de memoria al momento de escalar los datos. Sin embargo, es algo factible desde el punto de vista estrictamente funcional.

```python
cols = ["Sex", "Season", "Medal", "Year", "Age", "Height", "Weight"]  # columnas a escalar

pf.scale(
    columns=cols,  # columnas a escalar
    num_func_transformer=np.log,  # funci√≥n de transformaci√≥n de las variables num√©ricas
    num_scaler=MinMaxScaler(),  # scaler de las variables num√©ricas
    cat_encoder=OneHotEncoder(handle_unknown="ignore"),  # encoder de las variables categ√≥ricas
)
```

### 3. Creaci√≥n de Clusters y Anomal√≠as

En primer lugar analizaremos el apartado de clustering, para esto utilizaremos el m√©todo `make_clusters` de la clase `Profiler`. Este m√©todo se encarga de generar clusters de los datos usando alg√∫n algoritmo de clusterizaci√≥n. En este caso, se utilizar√° el algoritmo `KMeans` de `sklearn` y se generar√° un estudio del codo donde se√±ale la cantidad de clusters optimos para el desarrollo. Adem√°s, se proyectar√°n los datos a 2 dimensiones usando la t√©cnica de reducci√≥n de dimensionalidad `PCA` y se graficar√°n los resultados coloreando por cluster. Cabe mencionar que la elecci√≥n del n√∫mero de clusters se hace visualizando el gr√°fico del codo, introduciendo el √≥ptimo a trav√©s de la interfaz `input()`, esto se hace con el fin de que el usuario pueda elegir el n√∫mero de clusters que considere √≥ptimo.

```python
cols = pf.scaled_df.columns.tolist()  # columnas a analizar

pf.make_clusters(
    columns=cols,  # columnas a analizar
    clustering_model=KMeans(),  # KMeans
    dim_reduc_model=PCA(n_components=2, random_state=42),  # PCA
    max_clusters=10,  # n√∫mero m√°ximo de clusters a generar
    df="scaled",  # original, clean, scaled
    random_state=42,  # semilla para reproducibilidad
)
```

Ahora analizaremos el apartado de detecci√≥n de anomal√≠as, para esto utilizaremos el m√©todo `detect_anomalies` de la clase `Profiler`. Este m√©todo se encarga de detectar anomal√≠as en los datos. En este caso, se utilizar√° el algoritmo `IsolationForest` de `sklearn` y se proyectar√°n los datos a 2 dimensiones usando la t√©cnica de reducci√≥n de dimensionalidad `PCA` y se graficar√°n los resultados coloreando por las etiquetas predichas por el detector de anomal√≠as.

```python
pf.detect_anomalies(
    columns=cols,  # columnas a analizar
    anomaly_detector=IsolationForest(),  # IsolationForest
    dim_reduc_model=PCA(n_components=2, random_state=42),  # PCA
    contamination="auto",  # porcentaje de anomal√≠as esperadas
    df="scaled",  # original, clean, scaled
    random_state=42,  # semilla para reproducibilidad
)
```
