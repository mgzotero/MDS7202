Responda las siguientes preguntas (0.2 pts c/u):


**1. ¿Qué es `Git` y cómo se relaciona con el control de versiones?**
   
Un control de versiones es un sistema que registra los cambios realizados en un archivo o conjunto de archivos a lo largo del tiempo, de modo que se puedan ver *snapshots* (o instantáneas) antiguas de un proyecto, ver los cambios que han ocurrido en el mismo (con sus descripciones de por qué se hicieron estos cambios), trabajar colaborativamente en paralelo, etc. Existen sistemas de control de versiones (llamados VCS) locales, centralizados y distribuidos. Git es un sistema de control de versiones distribuido, lo que significa que cada usuario tiene una copia local del repositorio en su totalidad, y que los cambios se sincronizan entre los usuarios a través de un servidor remoto.


**2. Nombre y explique 3 diferencias entre `Git` y `Github`.**
   
| Git                                                                                                | GitHub                                                               |
|----------------------------------------------------------------------------------------------------|----------------------------------------------------------------------|
| 1. Es un sistema de control de versiones distribuido                                               | 1. Es un servicio de *hosting* de repositorios basado en web         |
| 2. Utiliza la línea de comandos para ejecutar comandos de control de versiones en los repositorios | 2. Proporciona una GUI intuitiva para la gestión de los repositorios |
| 3. Crea un repositorio local para *trackear* los cambios locales                                   | 3. Es un espacio para subir una copia de los repositorios de Git     |


**3. ¿Qué es un `hash` y cómo se representa en un `commit`? De un ejemplo.**

Un *hash* en Git es un código único y fijo (basado en SHA-1) que se calcula a partir del contenido de un objeto, como un árbol, un archivo o un commit. Sirve para identificar ese objeto de manera única. En un commit, el hash se usa para identificar el commit mismo y también se usa en las ramas y etiquetas para apuntar a commits específicos. Por ejemplo, un commit puede tener un hash como **a1b2c3d4e5f6**, que se crea a partir de su contenido y detalles. Este **hash** es esencial para seguir, identificar y gestionar cambios en un repositorio.


**4. ¿Que elementos componen a la estructura de datos de `Git`?**

Dentro de este contexto, se encuentran tres elementos esenciales: los *trees*, los *blobs* y los *commits*. Los *trees* desempeñan un papel similar al de los directorios, ya que albergan referencias cruzadas con otros *trees* y *blobs*. Por su parte, los *blobs* funcionan como los archivos propiamente dichos, ya que almacenan el contenido real de dichos archivos. En cuanto a los *commits*, asumen la función de instantáneas (*snapshots*) de un proyecto en un momento específico. Estos commits incluyen enlaces a los *trees* y *blobs* que configuran el proyecto, así como metadatos como el autor, el mensaje, el *hash* del commit previo, entre otros detalles relevantes.


**5. ¿Que comandos de `Git` sirven para crear un repositorio y hacer seguimiento a cambios en los archivos?**

- Crear un repositorio: `git init`
- Clonear un repositorio: `git clone url_repo_remoto`
- Vincular un repositorio local con uno remoto: `git remote add nombre_repo_remoto url_repo_remoto`
- Recopilar el estado del repositorio remoto: `git fetch nombre_repo_remoto`
- Obtener el estado del repositorio remoto: `git pull nombre_repo_remoto nombre_branch`
- Agregar archivos al *staging area*: `git add directorio_del_archivo` (o `git add .` para agregar todos los archivos del directorio actual)
- Confirmar cambios: `git commit -m "Comentario descriptivo del commit"`
- Enviar cambios al repositorio remoto: `git push nombre_repo_remoto nombre_branch`
- Hacer seguimiento a cambios en los archivos: `git status`
- Revisar las diferencias entre el último commit y el workspace actual: `git diff`
- Revisar el registro de commits: `git log`

Responda si las siguientes afirmaciones son Verdaderas (V) o Falsas (F). Justifique en **cualquiera** de los casos (0.2 pts c/u):


6. `Git` siempre resuelve automáticamente los problemas de lineas al hacer merge entre diferentes ramas.

**Falso**. Git no siempre resuelve automáticamente los problemas de líneas al hacer merge entre diferentes ramas. Puede surgir un conflicto cuando dos ramas modifican las mismas líneas en un archivo. Git requerirá que el usuario resuelva manualmente estos conflictos antes de completar el merge.


7. Ejecutar `git pull` sobre un repositorio es equivalente a ejecutar `git clone`.

**Falso**. Ejecutar `git pull` sobre un repositorio es equivalente a ejecutar `git fetch` y luego `git merge`, es decir, se obtiene el estado del repositorio remoto y se mezcla con el estado del repositorio local. Por otro lado, `git clone` crea una copia completa del repositorio remoto en la máquina local.


8. `git status` genera los mismos resultados que `git log`.

**Falso**. El comando `git status` muestra el estado actual de los archivos en el workspace y en el área de stage, mientras que `git log` muestra el registro de commits del repositorio.


9.  En `Git` es posible deshacer cambios y volver a versiones anteriores del código.

**Verdadero**. En Git es posible deshacer cambios y volver a versiones anteriores del código. Para ello, se puede utilizar el comando `git reset` para volver a un commit anterior, `git revert` para deshacer un commit específico o `git checkout` para volver temporalmente a un commit específico.


10. El comando `git commit` se utiliza para enviar cambios al repositorio remoto.

**Falso**. El comando `git commit` se utiliza para guardar los cambios en el repositorio local. Para enviar los cambios al repositorio remoto se debe utilizar el comando `git push`.
