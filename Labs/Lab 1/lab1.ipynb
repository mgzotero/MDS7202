**1. ¬øQu√© es `Git` y c√≥mo se relaciona con el control de versiones?**
   
Un control de versiones es un sistema que registra los cambios realizados en un archivo o conjunto de archivos a lo largo del tiempo, de modo que se puedan ver *snapshots* (o instant√°neas) antiguas de un proyecto, ver los cambios que han ocurrido en el mismo (con sus descripciones de por qu√© se hicieron estos cambios), trabajar colaborativamente en paralelo, etc. Existen sistemas de control de versiones (llamados VCS) locales, centralizados y distribuidos. Git es un sistema de control de versiones distribuido, lo que significa que cada usuario tiene una copia local del repositorio en su totalidad, y que los cambios se sincronizan entre los usuarios a trav√©s de un servidor remoto.


**2. Nombre y explique 3 diferencias entre `Git` y `Github`.**
   
| Git                                                                                                | GitHub                                                               |
|----------------------------------------------------------------------------------------------------|----------------------------------------------------------------------|
| 1. Es un sistema de control de versiones distribuido                                               | 1. Es un servicio de *hosting* de repositorios basado en web         |
| 2. Utiliza la l√≠nea de comandos para ejecutar comandos de control de versiones en los repositorios | 2. Proporciona una GUI intuitiva para la gesti√≥n de los repositorios |
| 3. Crea un repositorio local para *trackear* los cambios locales                                   | 3. Es un espacio para subir una copia de los repositorios de Git     |


**3. ¬øQu√© es un `hash` y c√≥mo se representa en un `commit`? De un ejemplo.**

Un *hash* en Git es un c√≥digo √∫nico y fijo (basado en SHA-1) que se calcula a partir del contenido de un objeto, como un √°rbol, un archivo o un commit. Sirve para identificar ese objeto de manera √∫nica. En un commit, el hash se usa para identificar el commit mismo y tambi√©n se usa en las ramas y etiquetas para apuntar a commits espec√≠ficos. Por ejemplo, un commit puede tener un hash como **a1b2c3d4e5f6**, que se crea a partir de su contenido y detalles. Este **hash** es esencial para seguir, identificar y gestionar cambios en un repositorio.


**4. ¬øQue elementos componen a la estructura de datos de `Git`?**

Dentro de este contexto, se encuentran tres elementos esenciales: los *trees*, los *blobs* y los *commits*. Los *trees* desempe√±an un papel similar al de los directorios, ya que albergan referencias cruzadas con otros *trees* y *blobs*. Por su parte, los *blobs* funcionan como los archivos propiamente dichos, ya que almacenan el contenido real de dichos archivos. En cuanto a los *commits*, asumen la funci√≥n de instant√°neas (*snapshots*) de un proyecto en un momento espec√≠fico. Estos commits incluyen enlaces a los *trees* y *blobs* que configuran el proyecto, as√≠ como metadatos como el autor, el mensaje, el *hash* del commit previo, entre otros detalles relevantes.


**5. ¬øQue comandos de `Git` sirven para crear un repositorio y hacer seguimiento a cambios en los archivos?**

- Crear un repositorio: `git init`
- Clonear un repositorio: `git clone url_repo_remoto`
- Vincular un repositorio local con uno remoto: `git remote add nombre_repo_remoto url_repo_remoto`
- Recopilar el estado del repositorio remoto: `git fetch nombre_repo_remoto`
- Obtener el estado del repositorio remoto: `git pull nombre_repo_remoto nombre_branch`
- Agregar archivos al *staging area*: `git add directorio_del_archivo` (o `git add .` para agregar todos los archivos del directorio actual)
- Confirmar cambios: `git commit -m "Comentario descriptivo del commit"`
- Enviar cambios al repositorio remoto: `git push nombre_repo_remoto nombre_branch`
- Hacer seguimiento a cambios en los archivos: `git status`
- Revisar las diferencias entre el √∫ltimo commit y el workspace actual: `git diff`
- Revisar el registro de commits: `git log`

**Falso**. Git no siempre resuelve autom√°ticamente los problemas de l√≠neas al hacer merge entre diferentes ramas. Puede surgir un conflicto cuando dos ramas modifican las mismas l√≠neas en un archivo. Git requerir√° que el usuario resuelva manualmente estos conflictos antes de completar el merge.

11. `git stash` (0.25 puntos)

`git stash` es un comando en Git que te permite temporalmente guardar los cambios no comprometidos en tu workspace, lo que te permite cambiar de branch o realizar otras operaciones sin tener que comprometer o descartar tus cambios. Esto es √∫til cuando se est√° trabajando en una feature o haciendo un bugfix y se necesita cambiar de contexto r√°pidamente sin hacer un commit.

    
12. `git reset`. ¬øCu√°l es la diferencia de este comando con `git revert`?. (0.25 puntos)

Tanto `git reset` como `git revert` son comandos de Git que se utilizan para deshacer cambios, pero lo hacen de manera diferente y en contextos distintos. `git reset` por su parte cambia la posici√≥n de la branch y el puntero de HEAD en el historial de commits, y puede descartar cambios. Por otro lado, `git revert` crea un nuevo commit que deshace los cambios de un commit espec√≠fico sin cambiar la historia de confirmaci√≥n existente. Es decir, `git reset` es m√°s dr√°stico y puede cambiar el historial de commits, mientras que `git revert` es m√°s seguro y crea nuevos commits para deshacer los cambios de commits anteriores sin alterar el historial.
